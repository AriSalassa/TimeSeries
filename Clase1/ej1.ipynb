Graficar 3 series de tiempo con los ejemplos del repositorio y describirlas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# YPF Dataset

ypf_dataset = pd.read_csv("../Datasets/YPFD.2000.2021.csv")
# Transform fechaHora column to datetime
ypf_dataset["fechaHora"] = pd.to_datetime(ypf_dataset["fechaHora"])
# Drop zero values
ypf_dataset = ypf_dataset[ypf_dataset.ultimoPrecio != 0]
ypf_dataset.head()
```

```python
ypf_dataset.describe()
```

```python
# Plot dataset
fig, ax = plt.subplots(figsize=(12, 12))

date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_form)
plt.grid()
plt.plot(ypf_dataset["fechaHora"], ypf_dataset["ultimoPrecio"])
```

En el gráfico anterior se muestra la variación del precio de la acción de YPF. Se puede observar un que el precio de la acción crece a un ritmo en promedio constante hasta el mes de Enero de 2020, donde alcanza un pico de 870 aproximadamente. En este estadío comienza un periódo a la baja que culmina con un derrumbe del precio de la acción a menos de 250 en el período de Abril de 2020. El valor de la acción se recupera rápidamente a valores previos al derrumbe. Para Agosto del 2020 el valor es de 800. Entre Agosto de 2020 y Mayo de 2021 se observa una oscilación del valor alrededor de los 680. Luego de ello se observa un período en alza que culima con un pico de 950 en Junio de 2021, momento a partir del cual pareciera que volvería a ocurrir periódo de oscilación alrededor de 800. No se observa una estacionalidad. 

```python
# Byma Dataset

byma_dataset = pd.read_csv("../Datasets/BYMA.csv")
# Transform fechaHora column to datetime
byma_dataset["fechaHora"] = pd.to_datetime(byma_dataset["fechaHora"])
# Drop zero values
byma_dataset = byma_dataset[byma_dataset.ultimoPrecio != 0]
byma_dataset.head()
```

En el gráfico anterior se muestra la variación del precio de la acción de BYMA. Se puede observar un que el precio de la acción crece a un ritmo en promedio constante dede Julio de 2017 hasta el mes de Abril de 2018, donde alcanza un pico de 425 aproximadamente. Durante un año el precio se matiente en promedio constante oscilando alrededor de 460. Luego del mes de Abril de 2019, el precio cae hasta un primer límite de 280 en Junio de 2019. Inmediatamente el precio de la acción recuperra el valor que tenía previo a la caída y se mantiene allí durante tres meses. En Septiembre de 2019 hay una abrupta caída del precio a un valor de 225, donde se mantiene otros tres meses. De Septiembre a Abril de 2020 se ve una recuperación del valor alrededor de los 280. Luego de ese momento la acción despega, creciendo en forma en promedio constante hasta llegar a un pico de 790 en el mes de Junio de 2021, período a parir del cual la proyección de la tendencia sería constante o a la alza. No se observa una estacionalidad.
