Graficar 3 series de tiempo con los ejemplos del repositorio y describirlas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# YPF Dataset

ypf_dataset = pd.read_csv("../Datasets/YPFD.2000.2021.csv")
# Transform fechaHora column to datetime
ypf_dataset["fechaHora"] = pd.to_datetime(ypf_dataset["fechaHora"])
# Drop zero values
ypf_dataset = ypf_dataset[ypf_dataset.ultimoPrecio != 0]
ypf_dataset.head()
```

```python
ypf_dataset.describe()
```

```python
# Plot dataset
fig, ax = plt.subplots(figsize=(12, 12))

date_form = DateFormatter("%Y-%m-%d")
ax.xaxis.set_major_formatter(date_form)
plt.grid()
plt.plot(ypf_dataset["fechaHora"], ypf_dataset["ultimoPrecio"])
```

En el grfico anterior se muestra la variacin del precio de la accin de YPF. Se puede observar un que el precio de la accin crece a un ritmo en promedio constante hasta el mes de Enero de 2020, donde alcanza un pico de 870 aproximadamente. En este estado comienza un perido a la baja que culmina con un derrumbe del precio de la accin a menos de 250 en el perodo de Abril de 2020. El valor de la accin se recupera rpidamente a valores previos al derrumbe. Para Agosto del 2020 el valor es de 800. Entre Agosto de 2020 y Mayo de 2021 se observa una oscilacin del valor alrededor de los 680. Luego de ello se observa un perodo en alza que culima con un pico de 950 en Junio de 2021, momento a partir del cual pareciera que volvera a ocurrir perido de oscilacin alrededor de 800. No se observa una estacionalidad.
